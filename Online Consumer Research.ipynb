## Abstract

This paper is a student project dedicated to the application of ACBC in the context of gaming consoles industry. The main product solution was presented in the form of Xbox Series S/X consoles. This paper describes the full scope of steps required for the application of the method in the industry. Using the sample of 120 respondents, results on specific metrics of customers’ preferences were acquired. Thus, in terms of recommendations to the customer in the context of the case role-play setting, study presumes to focus on the lower price point for the consoles. In addition, three the most efficient bundles, WTP (willingness to pay), shares of preference metrics were derived with the help of the academic version of Sawtooth software.

## Attributes and Attributes Levels

## 1. Console Type

The first attribute is the console type. Consoles are the main attribute within our list of attributes. Xbox provides two Xbox console families: Series S and Series X.  Xbox Series S is an entry-level console suitable for the majority of customers because it provides substantial value at a lower price point. In contrast, Xbox Series X provides the same features as Series S plus additional features such as a significantly higher picture quality due to a processor that is approximately three times faster than that of Series S and a Blue-Ray disk drive, both of which contribute to its higher price point (Xbox, n.d.a).

Xbox Series S is available in two versions differentiated by their storage capacity: 512GB and 1TB, while Xbox Series X has one version that is 1TB (Xbox, n.d.b) We believe that the existing product line needs to be revised. Therefore, we propose a new hypothetical product, Xbox Series X Digital Edition 1TB. The decision to add this product was based on the following assumptions: Firstly, the significantly lower price of Xbox Series S compared to Series X combined with ambiguous product differentiation and not evident utility gain, may lead consumers to prefer the cheaper console. This hypothesis is supported by sales disproportion provided by multiple gaming magazines which indicate that “74.8% of Xbox owners own the Series S, while the Series X is only 25.1%.”(Makar, 2023)
Secondly, the main competitor of the current generation of Microsoft Xbox is Sony PlayStation. PlayStation has a similar product to what we are introducing, PlayStation 5 Digital Edition, which shows that there is an existing market and demand for such a product that we as Xbox can explore (PlayStation, n.d.).

On this basis, we conclude that adding a hypothetical alternative in the form of a console that would take the price position between Xbox Series S and Xbox Series X can be a fruitful endeavor. We have set four attribute levels for this attribute. Level 1 includes the Xbox Series S 512GB as the base option (+0.00€), Xbox Series S 1TB (+50,99€), Xbox Series X Digital Edition 1TB (+150,99€), and finally Xbox Series X 1TB (+240,99€). ## D-Efficiency

To evaluate whether our study design was good enough to actually get precise point estimates, we ran a test design with 249 robotic respondents using sawtooth software and these are the results we got:

1. Each respondent evaluated each attribute level at least 2 times which is the minimum recommended number of times each respondent should see a level in order to evaluate it.

2. The expected standard errors of each attribute level for part with utilities were below 0.05 except for the price, it was 0.09, which would be considered a limitation of our study design. 
Thus, we concluded that our study design is good enough to actually get precise point estimates.

## HB Analysis

In order to reach our final model, we followed these steps:

1. We ran a counts analysis on our final data using the analysis tab in Sawtooth software. We got the summed price frequencies in the resulted file of our counts analysis. We put those summed price frequencies into an R code which gave us 11 price breakpoints of our data. We used those 11 price breakpoints to run a piecewise HB estimation in Sawtooth software. In the resulted file of our piecewise HB estimation, we got the average utilities in the output file.

2. We ran a scatter plot on R to see the relationship between the price breakpoints and the utilities. The relationship was linear with an R squared value of 0.99. Thus, we decided to proceed with HB linear price coding.

```r
pacman::p_load("DescTools",
               "readxl",
               "tidyverse")

df <- read_xlsx("Piecewise n = 6 Report.xlsx", sheet = "Summary")

colnames(df) <- c("label", "util", "sd")

df <- df[-c(1:which(str_detect(df$label, "Average Utilities")), (which(str_detect(df$label, "Average Importances")) - 1):nrow(df)), ]

df <- df[, -3]

df <- df %>% 
dplyr::filter(as.integer(str_detect(.$label, "PRICE")) == 1)

colnames(df) <- c("price","utility")

df$price <- gsub("PRICE: ", "", df$price)

df$price <- as.numeric(df$price)
df$utility <- as.numeric(df$utility)

plot(df$price[2:10], df$utility[2:10],
     main = "Price function",
     xlab = "Price ", ylab = "Utilitiy ", pch = 19, ylim = c(-300, 200)
) + abline(lm(utility ~ price, data = df[2:10, ]), col = "red")

round(summary(lm(utility ~ price, data = df[2:10, ]))$r.squared, digits = 3)
```

3. We ran an HB estimation with linear price coding and constraining price negatively.

4. We ran an interaction search to see if there are interactions in our model. There were two interactions (additional storage with price, online subscription with price)

5. We ran additional 3 HB estimations with linear price coding since we have interactions; with the first interaction, with the second interaction, with both interactions at the same time.

6. We compared the predictive validity of the four models we have, and we decided to proceed with the linear price model with no interactions since the predictive validity of this model is very similar to those with interactions and to avoid complexity in terms of interpreting the main effects of our model.

```r
if (!"pacman" %in% rownames(installed.packages())) install.packages("pacman")

library(pacman)
pacman::p_load(
  "readxl",
  "tidyverse",
  "psych",
  "irr"
)

df <- readxl::read_xlsx("Simulation Report Add Storage  Online Sub ~ Price HOT.xlsx", sheet = "HOT | Individual Results")
head(df)
df <- df %>%
  filter(!row_number() %in% c(1, 2)) %>% # this will delete the first two rows
  mutate_all(., as.numeric) %>% # changes the data type to numeric (all except for 1st column = ID)
  rename_all(., ~ c("id", paste0("Option_", c(1:(ncol(df) - 1))))) %>% # rename all columns (id = id; all choices to Option_)
  mutate(across(2:ncol(.), ~ .x * 100)) # finally multiply by 100 to get percentages again
main_survey <- read.csv("Our good data.csv")

df <- df %>%
  merge(
    x = .,
    y = main_survey[c("sys_RespNum", "HOT")],
    by = 1
  ) %>%
  relocate(HOT, .after = id)

df %>%
  mutate(pred_choice = max.col(.[, c(3:ncol(.))])) %>% # this will determine the highest value among the alternatives
  summarise(hitrate = 100 * mean(HOT == pred_choice)) %>% # this first counts the number of times the actual choice equals the predicted choice and afterwards multiplies the result with 100
  round(3) # finally we round the results to 3 digits

no_cor <- df %>%
  mutate(pred_choice = max.col(.[, c(3:ncol(.))])) %>% # this will determine the highest value among the alternatives
  summarise(cor = sum(HOT == pred_choice)) %>% #  counts the number of times the actual choice equals the predicted choice
  unlist(.) %>% # stores the result as vector
  unname(.) # unnames the vector

binom.test(
  x = no_cor,
  n = nrow(df),
  p = 1 / (ncol(df) - 2)
)[["p.value"]] %>%
  round(3)

mae <- data.frame(Option = NA, pred_share = NA, obs_share = NA)

for (k in 3:ncol(df)) {
  mae[(k - 2), "Option"] <- k - 2
  
  mae[(k - 2), "pred_share"] <- mean(df[, k])
  
  mae[(k - 2), "obs_share"] <- sum(df$HOT == (k - 2)) / nrow(df) * 100
}

mae %>%
  summarise(mae = mean(abs(pred_share - obs_share))) %>%
  round(3) # round the results to 3 digits

mae %>% 
  rename("Predicted" = 2,
         "Observed" = 3) %>% 
  pivot_longer(data = ., 
               cols = -Option,
               names_to = "Condition",
               values_to = "Share") %>% 
  mutate(Option = paste0("Option ", Option)) %>% 
  ggplot(., aes(fill = Condition, y = Share, x = Option)) +
  geom_bar(width = 0.7, position = position_dodge(width = 0.8), stat = "identity") +
  theme_minimal() +
  ylab("Observed and prediected choice shares of options in HOT") +
  theme(
    legend.title = element_text(size = 13),
    legend.text = element_text(size = 12),
    axis.text.y = element_text(size = 12),
    axis.text.x = element_text(size = 12),
    axis.line = element_blank(),
    axis.title.y = element_text(size = 11),
    axis.title.x = element_blank(),
    legend.position = "bottom"
  ) +
  geom_text(
    aes(label = format(round(Share, 1), nsmall = 1)),
    position = position_dodge(width = 0.8),
    size = 3.5,
    vjust = -0.48
  ) +
  scale_y_continuous(limits = c(0, 100)) +
  scale_fill_manual("Legend",values = c("Predicted" = "#FF6666", "Observed" = "#D60060"))

mae %>%
  summarise(medae = median(abs(pred_share - obs_share))) %>%
  round(3) # round the results to 3 digits

mae %>%
  summarise(rmse = sqrt(mean((abs(pred_share - obs_share))^2))) %>%
  round(3) # round the results to 3 digits

opts <- c(
  "Option_1", "Option_2", "Option_3", "Option_4",
  "Option_5", "Option_6"
)

df2 <- df %>%
  mutate(pred = max.col(.[c(opts)]))


psych::cohen.kappa(
  x = select(df2, HOT, pred),
  alpha = .05
)$confid[1, ]

irr::kappa2(ratings = select(df2, HOT, pred))[["statistic"]] %>%
  round(3)

irr::kappa2(ratings = select(df2, HOT, pred))[["p.value"]] %>%
  round(3)

df$mhp <- 0

for (i in 1:nrow(df)) {
  df[i, "mhp"] <- df[i, (df[i, "HOT"] + 2)]
}

df %>%
  reframe(mhp = mean(mhp)) %>%
  round(3)
```

## Market Simulation

We conducted a product optimization analysis using Share of Preference and Genetic search settings.

All attribute levels were set on range resulting in a 58.6% “none” share of preferences (table 16). After analyzing the attribute levels of the best three bundles, it was decided to run another analysis by adding some limitations on the attribute levels. This decision was based on the observation that some of the attribute levels within the best three bundles in the result were not the attribute levels with the highest utilities, and we believed that adding some limitations on some attribute levels to have some control over the outcome could be beneficial in our case. Thus, we decided to run another analysis by adding the following limitations. Firstly, we excluded all extra storage capacity cards because we knew they did not have high preference. Secondly, we limited body color to “none” or grey color, since these two options had the highest results. Thirdly, we limited controllers by excluding the wired controller since we knew it decreased the choice likelihood of a bundle. Fourthly, we excluded the headphones attribute as a whole since we believed it caused an unjustifiable price increase with no added value. Fifthly, we limited game bundle 1 to FC24, Call of Duty or none option, and bundle 2 to Assassin's Creed, Read Dead Redemption II or “none” option. Moreover, to compare the current market setting with the three available consoles and the new hypothetical option, we limited the console to include only the Xbox Series S 512GB, Xbox Series S 1TB, and the Xbox Series X 1TB. As anticipated the result came better with 43.2% “none” share of preference (table 17). Consequently, we repeated the same analysis with the same limitations, however, we replaced the Xbox Series X 1TB with its most similar and new hypothetical counterpart the Xbox Series Digital Edition 1TB to see if it would result in a decrease in the “none” share of preference. Contrary to belief, it resulted in a worse result of 47.6% of the “None” option share of preference.  Moreover, the Xbox Digital edition 1TB was not even included among the 3 best bundles from this analysis (table 18).

Finally, after running all these analyses and comparing their results we decided that the three bundles in the second analysis were the best we could offer to the market.

## Limitations

Selection bias: Non-random sampling based on availability.
we recruited respondents from our friends, casual acquaintances and family members. Those were the people we were able to reach and convince them to fill in our survey without any monetary compensation.

Market simulation: affected by the lack of competitor data.
we didn’t include any data related to competitors when we did the market simulation. Market simulation was only based on the data and the insights we got from our study.

Bundle optimization: unable to optimize for profits and costs due to data unavailability.
 We didn’t include any data related to costs or profits when doing the product line optimization. We only did our product line optimization based on the data and the insights we got from our study.

Model complexity: excluded interaction models to balance complexity and fit.

Segmentation analysis: no significant results obtained. We tried to segment our respondents into different segments based on their answers to the questions we had in our study, however, we could not come up with clearly defined segments with significantly different behaviors in terms of answering the survey questions

Study design: standard error for price utility is 0.09, while most D-efficiency scores are below 0.9.

Lack of interest: 73 respondents have no interest in purchasing an Xbox console within the next 12 months.

Geographic limitation: 75% of respondents are not from Germany. The majority of the data we have and the interpretations we make are not based on data from potential customers in Germany.

Income limitation: 80% of the respondents have a monthly income below 2000 euros, which limits the extrapolation of our interpretations for higher income groups.
